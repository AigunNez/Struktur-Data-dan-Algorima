# Simple Queue ( Linier Queue ) -> Array

```python
class ArrayQueue:
    def __init__(self, maks):
        self.data = []
        self.front = 0
        self.rear = 0
        self.maks = maks
        
    def __len__(self):
        return self.rear
    
    def isEmpty(self):
        ## Cara Pertama
        # if self.rear == 0:
        #     return True
        # else:
        #     return False
        
        ## Cara Kedua
        # if self.rear == 0:
        #     return True
        # return False
        
        ## Cara Ketiga
        return self.rear == 0
    
    def display(self):
        if self.isEmpty():
            print("Queue is empty")
            return None
        else :
            print(self.data)
            
    def enqueue(self, new_data):
        if self.rear == self.maks:
            print("Queue is full")
            return None
        else :
            self.data.append(new_data)
            self.rear += 1
            
    def dequeue(self):
        if self.isEmpty():
            print("Queue is empty")
            return None
        else :
            self.data.pop(self.front)
            self.rear -= 1
            
    def peekFront(self):
        if self.isEmpty():
            print("Queue is empty")
            return None
        else :
            return self.data[self.front]
        
    def peekRear(self):
        if self.isEmpty():
            print("Queue is empty")
            return None
        else :
            return self.data[self.rear - 1]
```

```python
maks_queue = 5
queue = ArrayQueue(maks_queue)
queue.display()

queue.enqueue(1)
queue.enqueue(2)
queue.display()

queue.enqueue(3)
queue.enqueue(4)
queue.enqueue(5)
queue.display()

queue.dequeue()
queue.display()

print("Data paling depan adalah", queue.peekFront())
print("Data paling belakang adalah", queue.peekRear())
```

# Circular Queue ( Array )

```python
class CircQueue:
    def __init__(self, maks):
        self.data = [None] * maks
        # self.data = [1, 2, 3, 4, 5]
        self.front = 0
        self.rear = 0
        self.maks = maks
        self.count = 0
        
    def __len__(self):
        return self.rear
    
    def isEmpty(self):
        ## Cara Pertama
        # if self.rear == 0:
        #     return True
        # else:
        #     return False
        
        ## Cara Kedua
        # if self.rear == 0:
        #     return True
        # return False
        
        ## Cara Ketiga
        return self.rear == 0
    
    def display(self):
        if self.isEmpty():
            print("Queue is empty")
            return None
        else :
            print("Isi Queue:", end=" ")
            index = self.front
            for i in range(self.count):
                print(self.data[index], end=" ")
                index = (index + 1) % self.maks
                
            print()
            
    def enqueue(self, new_data):
        if self.count == self.maks:
            print("Queue is full")
            return None
        else :
            self.data[self.rear] = new_data
            self.rear = (self.rear + 1) % self.maks
            self.count += 1
            
    def dequeue(self):
        if self.isEmpty():
            print("Queue is empty")
            return None
        else :
            self.data[self.front] = None
            self.front = (self.front + 1) % self.maks
            self.count -= 1
```

```python
maks_queue = 5
circ_queue = CircQueue(maks_queue)
circ_queue.display()

circ_queue.enqueue(1)
circ_queue.enqueue(2)
circ_queue.display()

circ_queue.enqueue(3)
circ_queue.enqueue(4)
circ_queue.display()

circ_queue.dequeue()
circ_queue.display()
```

# Simple Queue ( Linear Queue ) -> Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class LinQueue:
    def __init__(self):
        self.front = None
        self.rear = None
        
    def isEmpty(self):
        ## Cara Pertama
        if self.front is None:
            return True
        else:
            return False
        
        ## Cara Kedua
        # return self.front is None
        
    def display(self):
        if self.isEmpty():
            print("Queue is empty")
            return None
        else :
            temp = self.front
            print("Isi Queue:", end=" ")
            while temp:
                print(temp.data, end=" ")
                temp = temp.next
            print()
            
    def enqueue(self, new_data):
        new_node = Node(new_data)
        if self.isEmpty():
            ## Cara Pertama
            self.front = new_node
            self.rear = new_node
            
            ## Cara Kedua
            # self.front = self.rear = new_node
        else :
            self.rear.next = new_node
            self.rear = new_node
            
    def dequeue(self):
        if self.isEmpty():
            print("Queue is empty")
            return None
        else :
            removed = self.front.data
            self.front = self.front.next
            if self.front is None:
                self.rear = None
                
            return removed
```

```python
lin_queue = LinQueue()
lin_queue.display()

lin_queue.enqueue(1)
lin_queue.enqueue(2)
lin_queue.display()

lin_queue.enqueue(3)
lin_queue.enqueue(4)
lin_queue.display()

print("Data yang dihapus = ", lin_queue.dequeue())
lin_queue.display()

print("Data yang dihapus = ", lin_queue.dequeue())
lin_queue.display()
```

# Circular Queue (Linked List)

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class CirLinQueue():
    def __init__(self):
        self.front = None
        self.rear = None
        
    def isEmpty(self):
        ## Cara Pertama
        if self.front is None:
            return True
        else:
            return False
        
        ## Cara Kedua
        # return self.front is None
        
    def display(self):
        if self.isEmpty():
            print("Queue is empty")
            return None
        else :
            temp = self.front
            print("Isi Queue:", end=" ")
            while temp:
                print(temp.data, end=" ")
                temp = temp.next
                if temp == self.front:
                    break
                
            print()
            
    def enqueue(self, new_data):
        new_node = Node(new_data)
        if self.isEmpty():
            ## Cara Pertama
            self.front = new_node
            self.rear = new_node
            
            ## Cara Kedua
            # self.front = self.rear = new_node
        else :
            new_node.next = self.front
            self.rear.next = new_node
            self.rear = new_node
            
    def dequeue(self):
        if self.isEmpty():
            print("Queue is empty")
            return None
        else :
            removed = self.front.data
            
            if self.front == self.rear:
                self.rear = None
                self.front = None
            else :
                self.front = self.front.next
                self.rear.next = self.front
                
            return removed
```

```python
circ_lin_queue = CirLinQueue()
circ_lin_queue.display()

circ_lin_queue.enqueue(1)
circ_lin_queue.enqueue(2)
circ_lin_queue.display()

circ_lin_queue.enqueue(3)
circ_lin_queue.enqueue(4)
circ_lin_queue.display()

print("Data yang dihapus :", circ_lin_queue.dequeue())
circ_lin_queue.display()

print("Data yang dihapus :", circ_lin_queue.dequeue())
circ_lin_queue.display()
```
