# Bubble Sort

```python
def bubbleSort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]: # Ascending order
            # if arr[j] < arr[j+1]: # Descending order
                # data = arr[j]
                # arr[j] = arr[j+1]
                # arr[j+1] = data
                
                arr[j], arr[j+1] = arr[j+1], arr[j]
                
data = [64, 34, 25, 12, 22, 11, 90]
print("Unsorted array is:", data)
bubbleSort(data)
print("Sorted array is:", data)
```

# Selection Sort

```python
def selectionSort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i+1, n):
            if arr[j] < arr[min_index]: # Ascending order
            # if arr[j] > arr[min_index]: # Descending order
                min_index = j
                
        arr[i], arr[min_index] = arr[min_index], arr[i]
        
data = [64, 25, 12, 22, 11]
print("Unsorted array is:", data)
selectionSort(data)
print("Sorted array is:", data)
```

# Insertion Sort

```python
def insertionSort(arr):
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        # while j >= 0 and key > arr[j]: # Descending order
        while j >= 0 and key < arr[j]: # Ascending order
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
        
data = [12, 11, 13, 5, 6]
print("Unsorted array is:", data)
insertionSort(data)
print("Sorted array is:", data)
```

# Merge Sort

```python
def mergeSort(arr):
    n = len(arr)
    if n > 1:
        mid = n // 2
        left = arr[:mid]
        right = arr[mid:]
        
        mergeSort(left)
        mergeSort(right)
        
        i = j = k = 0
        
        # Menggabungkan dua sub-array
        while i < len(left) and j < len(right):
            # if left[i] < right[j]: # Ascending order
            if left[i] > right[j]: # Descending order
                arr[k] = left[i]
                i += 1
            else :
                arr[k] = right[j]
                j += 1
                
            k += 1
            
        while i < len(left):
            arr[k] = left[i]
            i += 1
            k += 1
            
        while j < len(right):
            arr[k] = right[j]
            j += 1
            k += 1
            
data = [38, 27, 43, 3, 9, 82, 10]
print("Unsorted array is:", data)
mergeSort(data)
print("Sorted array is:", data)
```

# Quick Sort


```python
def quickSort(arr):
    n = len(arr)
    if n <= 1:
        return arr
    else :
        pivot = arr[0]
        
        less = []
        for i in range(1, n):
            if arr[i] <= pivot: # Ascending order
            # if arr[i] >= pivot: # Descending order
                less.append(arr[i])
        
        greater = []
        for i in range(1, n):
            if arr[i] > pivot: # Ascending order
            # if arr[i] < pivot: # Descending order
                greater.append(arr[i])
        
        # less = [x for x in arr[1:] if x <= pivot]
        # greater = [x for x in arr[1:] if x > pivot]
        return quickSort(less) + [pivot] + quickSort(greater)
    
data = [90, 7, 8, 50, 0, 3]
print("Unsorted array is:", data)
sorted_data = quickSort(data)
print("Sorted array is:", sorted_data)
```

# Bucket Sort


```python
def insertionSort(arr):
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and key > arr[j]: # Descending order
        # while j >= 0 and key < arr[j]: # Ascending order
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
        
    return arr

def bucketSort(arr, bucket_size=5):
    bucket = [[] for _ in range(bucket_size)]
    
    for num in arr:
        index = int(num * bucket_size)
        bucket[index].append(num)
        
    for i in range(bucket_size):
        bucket[i] = insertionSort(bucket[i])
        
    result = []
    # for b in bucket: # Ascending order
    for b in reversed(bucket): # Descending order
        result.extend(b)
    return result

data = [0.78, 0.17, 0.39, 0.26, 0.72, 0.94, 0.21, 0.55]
print("Unsorted array is:", data)
sorted_data = bucketSort(data, 10)
print("Sorted array is:", sorted_data)
```
