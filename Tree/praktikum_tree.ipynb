# General Tree (Array)

```python
class GenTree:
    def __init__(self):
        self.data = []
        
    def add(self, data, parent_index):
        self.data.append([data, parent_index])
        
    def displayTree(self, index=0, indent=0):
        # print(self.data)
        print("  " * indent + self.data[index][0])
        for i in range(len(self.data)):
            if self.data[i][1] == index:
                self.displayTree(i, indent + 1)
                
    def getDescendants(self, index, type=None):
        # if type is None:
        #     descendants = [self.data[index][0]]
        # else:
        #     descendants = [index]
        descendants = [self.data[index][0]] if type is None else [index]
        for i in range(len(self.data)):
            if self.data[i][1] == index:
                # if type is None:
                #     descendants.append(self.data[i][0])
                # else:
                #     descendants.append(i)
                
                descendants.extend(self.getDescendants(i)) if type is None else descendants.append(i)
        return descendants
    
    def removeSubTree(self, index):
        descendants = sorted(self.getDescendants(index, type="remove"), reverse=True)
        for i in descendants:
            print(f"Menghapus node: {self.data[i][0]}")
            del self.data[i]
        
```

```python
tree = GenTree()
tree.add("A", -1)
tree.add("B", 0)
tree.add("C", 0)

tree.add("D", 1)
tree.add("E", 1)

tree.add("F", 2)
tree.add("G", 3)
tree.add("H", 3)

tree.displayTree()

print()
print("Descendants dari B")
print(tree.getDescendants(3))

print(tree.getDescendants(3, type="remove"))    

print()
print("Menghapus subtree D")
tree.removeSubTree(3)
tree.displayTree()
```

# Binary Tree

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        
class BinaryTree:
    def __init__(self, root):
        self.root = Node(root)
        
    def insertLeft(self, current_node, new_data):
        if current_node.left is None:
            current_node.left = Node(new_data)
        else :
            new_node = Node(new_data)
            new_node.left = current_node.left
            current_node.left = new_node
            
    def insertRight(self, current_node, new_data):
        if current_node.right is None:
            current_node.right = Node(new_data)
        else:
            new_node = Node(new_data)
            new_node.right = current_node.right
            current_node.right = new_node

    def getRoot(self):
        return self.root.data
    
    def getLeft(self, current_node):
        # if current_node.left is None:
        #     return None
        # else:
        #     return current_node.left.data
        
        return current_node.left.data if current_node.left else None
        
    def getRight(self, current_node):
        # if current_node.right is None:
        #     return None
        # else:
        #     return current_node.right.data
        
        return current_node.right.data if current_node.right else None
    
    def deleteLeft(self, current_node):
        current_node.left = None
        
    def deleteRight(self, current_node):
        current_node.right = None
        
    def preOrder(self, node):
        if node:
            print(node.data, end=" ")
            self.preOrder(node.left)
            self.preOrder(node.right)
            
    def inOrder(self, node):
        if node:
            self.inOrder(node.left)
            print(node.data, end=" ")
            self.inOrder(node.right)
            
    def postOrder(self, node):
        if node:
            self.postOrder(node.left)
            self.postOrder(node.right)
            print(node.data, end=" ")
    
    def printTree(self, node, level=0):
        if node is not None:
            self.printTree(node.right, level + 1)
            print(" "* 4 * level + "->", node.data)
            self.printTree(node.left, level + 1)
```

```python
tree = BinaryTree("A")
tree.insertLeft(tree.root, "B")
tree.insertRight(tree.root, "C")
tree.insertLeft(tree.root.left, "D")
tree.insertRight(tree.root.left, "E")
tree.insertLeft(tree.root.right, "F")
# tree.insertRight(tree.root.right, "G")

print(f"Root dari tree adalah:", tree.getRoot())

print("Binary Tree:")
tree.printTree(tree.root)

print("Preorder Traversal:")
tree.preOrder(tree.root)

print("\nInorder Traversal:")
tree.inOrder(tree.root)

print("\nPostorder Traversal:")
tree.postOrder(tree.root)
```

```python
print(f"Left child dari root adalah:", tree.getLeft(tree.root)) 
print(f"Right child dari root adalah:", tree.getRight(tree.root))
# print(f"Left child dari left child adalah:", tree.getLeft(tree.root.left))

print("Binary Tree:")
tree.printTree(tree.root)

print("Menghapus left child dari root")
tree.deleteLeft(tree.root.left)
tree.deleteRight(tree.root.left)
tree.printTree(tree.root)
```
